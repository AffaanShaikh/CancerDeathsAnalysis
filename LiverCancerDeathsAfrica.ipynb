```python
#selecting top 10 countries where death due to liver cancer is more than 2000
#and keeping only relevant columns inplace and restting index
df_liver = df_africa.loc[df_africa['Liver cancer'] > 2000]
df_liver = df_liver[['Country','Year','Liver cancer']]
df_liver.reset_index(drop=True, inplace=True)
df_liver.head()
```

```python
#assigning Year as rows, Countries as columns and the deaths as values
sorted_liver = df_liver.pivot(index='Year', columns='Country', values='Liver cancer')
sorted_liver
```

```python
#plotting the final product of liver deaths over years in African countries
plt.figure(figsize=(10, 6))
for country in sorted_liver.columns:
    plt.plot(sorted_liver.index, sorted_liver[country], label=country)

plt.title('Liver Cancer Deaths Over 20 Years in Africa')
plt.xlabel('Year')
plt.ylabel('Liver Cancer Deaths')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
